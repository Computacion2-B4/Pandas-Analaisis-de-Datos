```python
#2.-Cuantos registros hay por continente
import pandas as pd

# URL del archivo CSV
url = 'https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-data.csv?raw=true'

# Cargar los datos desde la URL
df = pd.read_csv(url)

# Contar registros por continente
registros_por_continente = df['continent'].value_counts().reset_index()
registros_por_continente.columns = ['continent', 'count']

# Mostrar el resultado
print(registros_por_continente)

```

```python
#7.-Qué país ha tomado las medidas más estrictas hasta ahora?
import pandas as pd

# URL del archivo CSV
url = 'https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-data.csv?raw=true'

# Cargar los datos desde la URL
df = pd.read_csv(url)

# Filtrar datos relevantes (ejemplo)
df_filtered = df[['location', 'stringency_index', 'total_cases_per_million', 'total_deaths_per_million']]

# Agrupar por país y obtener estadísticas relevantes
stats_by_country = df_filtered.groupby('location').mean()

# Ordenar por stringency_index para ver los países con mayores índices de restricción
stats_by_country_sorted = stats_by_country.sort_values(by='stringency_index', ascending=False)

# Mostrar los primeros países más restrictivos
print("Países con las medidas más estrictas:")
print(stats_by_country_sorted.head(10))

print("\nEl pais que ha tomado las medidas más estrictas es CHINA\n")
```

```python
#8.-Cual fue el país en alcanzar los 10.000 casos activos por primera vez?
import pandas as pd

# URL del archivo CSV
url = 'https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-data.csv?raw=true'

# Cargar los datos desde la URL
df = pd.read_csv(url)

# Convertir la columna de fecha a tipo datetime
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')

# Filtrar los datos para tener las columnas relevantes
df_filtered = df[['location', 'date', 'total_cases', 'total_cases_per_million', 'total_deaths', 'total_deaths_per_million', 'total_tests']]

# Agrupar por país y encontrar la primera fecha en la que se superaron los 10,000 casos activos
primeros_10000_casos_activos = df_filtered[df_filtered['total_cases'] >= 10000].groupby('location')['date'].min().reset_index()

# Ordenar por fecha para encontrar el primer país en alcanzar los 10,000 casos activos
primer_pais_10000_casos_activos = primeros_10000_casos_activos.sort_values(by='date').iloc[0]

# Mostrar el resultado
print(f"El primer país en alcanzar los 10,000 casos activos fue {primer_pais_10000_casos_activos['location']} el {primer_pais_10000_casos_activos['date'].strftime('%Y-%m-%d')}.")
```

   9.-<strong>Cuantas filas</strong> hay por país ?.

```python
#En este paso creamos un nuevo DATAFRAME solo con los paises de Perú y Guinea

import pandas as pd

# URL del archivo CSV
url = 'https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-data.csv?raw=true'

# Cargar los datos desde la URL
df = pd.read_csv(url)

# Filtrar los datos para los países de interés: Perú y Guinea
paises_interes = ['Peru', 'Guinea']
df_interes = df[df['location'].isin(paises_interes)]

# Mostrar el cuadro con la información de los países de interés
print(df_interes)
```

```python
#9.-Cuantas filas hay por país ?
# Contar filas por país
filas_por_pais = df_interes['location'].value_counts().reset_index()
filas_por_pais.columns = ['location', 'count']

# Mostrar el resultado
print(filas_por_pais)
```

```python
#10.-Cuál es la fecha máxima por país ?.
# Asegurarse de que la columna de fecha está en formato de fecha
df_interes['date'] = pd.to_datetime(df_interes['date'], format='%Y-%m-%d')

# Agrupar por 'location' y encontrar la fecha máxima
fechas_maximas = df_interes.groupby('location')['date'].max().reset_index()

# Mostrar el resultado
print(fechas_maximas)
```

```python
#11.-Cuál es la fecha mínima por país ?.
# Convertir la columna de fecha a tipo de fecha
df_interes['date'] = pd.to_datetime(df_interes['date'], format='%Y-%m-%d')

# Encontrar la fecha mínima por país
fecha_minima_por_pais = df_interes.groupby('location')['date'].min().reset_index()

# Mostrar el resultado
print(fecha_minima_por_pais)
```

```python
#12.-Cuantos registros hay por "continente"?
# Contar registros por continente
registros_por_continente = df_interes['continent'].value_counts().reset_index()
registros_por_continente.columns = ['continent', 'count']

# Mostrar el resultado
print(registros_por_continente)
```

```python
#13.-Cuantos nuevos casos hubo como máximo en una fecha reportada por cada país??
# Convertir la columna de fecha a tipo de fecha
df_interes['date'] = pd.to_datetime(df_interes['date'], format='%Y-%m-%d')

# Encontrar el máximo de nuevos casos reportados por país en una fecha
max_nuevos_casos = df_interes.groupby('location')['new_cases'].max().reset_index()

# Mostrar el resultado
print(max_nuevos_casos)
```
