<div class="alert alert-info">Con <strong>Pandas</strong> podemos verificar si faltan valores en el conjunto de datos solamente tenemos que implementar <code>pd.isnull()</code>. Implementando esta instrucción nos devuelve una matriz booleana, una matriz de <strong>verdadero</strong> para valores faltantes y <strong>falso</strong> para valores no perdidos.</div>

```python
#Verificar si hay datos nulos en el DataFrame
print('Datos nulos en el DataFrame: ')


#Imprimir nulos TRUE
notas2.isnull()
```

## <font color=#33A5FF>****2. Frecuencia Simple de Datos****<tfont>

<div class="alert alert-info"> Vamos hallar la <strong>frecuencia</strong> de notas de un archivo  <code>csv</code> damos como resultado la nota mas alta</div>

```python
#cargamos un archivo CSV separado por punto y coma
notas3 = pd.read_csv('/content/drive/MyDrive/data/notas.csv', delimiter=';')
```

```python
#nombres de columnas
notas3.columns.values
```

```python
#buscamos cuantos alumnos han sacado nota 20 en Promedio
#creamos nuevo DataFrame con notas promedio y frecuencias
tab = pd.crosstab(index=notas3['Nota3'], columns='Frecuencia')

tab.head(15)
```

```python
#localizar alumnos que han sacado 20 en su Nota3
fila = tab.loc[tab.index==20]

#variable
x = fila['Frecuencia']
x = int(x)

#Impresion
print('\t \t \t FELICITACIOÓN')
print("===========================================================")
print('\n')
print("%d alumnos han sacado en su Nota3 20, mis felicitaciones!!"%x)
print('\n\n\n')
```

#### Datos Nulos NaN en ingresos de PBI `gdp_per_capita`

```python
# Ensure 'location' is included in the DataFrame
numeric_df = df.select_dtypes(include=['float', 'int'])
numeric_df['location'] = df['location']

# Group by 'location' and calculate the mean
mean_df = numeric_df.groupby('location').mean()

# Extract the 'gdp_per_capita' column
gdp_per_capita = mean_df['gdp_per_capita']
print(gdp_per_capita)
```

```python
# Calcular el promedio de 'gdp_per_capita', ignorando los valores nulos
promedio_gdp = df['gdp_per_capita'].mean()

# Crear una nueva columna rellenando los valores nulos con el promedio
df['gdp_per_capita_not_nulls'] = df['gdp_per_capita'].fillna(promedio_gdp)

# Mostrar las primeras filas del DataFrame para verificar
print(df.head())
```

### Preguntas?

### Cuál es el país con el mayor life_expectancy?

```python
# Seleccionar solo las columnas numéricas y agregar 'location'
numeric_df = df.select_dtypes(include=['float', 'int'])
numeric_df['location'] = df['location']

# Agrupar por 'location' y calcular la media para cada grupo
mean_df = numeric_df.groupby('location').mean()

# Asegurarse de que 'life_expectancy' está en el DataFrame medio y es de tipo numérico
if 'life_expectancy' in mean_df.columns:
    # Ordenar 'life_expectancy' en orden descendente
    sorted_life_expectancy = mean_df['life_expectancy'].sort_values(ascending=False)
    # Imprimir los valores ordenados
    print(sorted_life_expectancy)
else:
    print("La columna 'life_expectancy' no se encuentra en el DataFrame numérico.")
```

### Qué país tiene la menor cantidad de fumadores masculinos?

```python
# Seleccionar solo las columnas numéricas
numeric_df = df.select_dtypes(include=['float', 'int'])

# Añadir la columna 'location' al DataFrame numérico
numeric_df['location'] = df['location']

# Agrupar por 'location' y calcular la media para cada grupo
mean_df = numeric_df.groupby('location').mean()

# Asegurarse de que 'male_smokers' está en el DataFrame medio y es de tipo numérico
if 'male_smokers' in mean_df.columns:
    # Ordenar 'male_smokers' en orden ascendente
    sorted_male_smokers = mean_df['male_smokers'].sort_values(ascending=True)
    # Imprimir los valores ordenados
    print(sorted_male_smokers)
else:
    print("La columna 'male_smokers' no se encuentra en el DataFrame numérico.")
```
